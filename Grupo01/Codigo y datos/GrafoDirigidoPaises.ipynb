```python
param = str(input("Introduce a√±o entre los siguientes: 2018, 1996 y 1975: "))
```

```python
import csv
#%precision 2
%matplotlib notebook
import matplotlib.pyplot as plt 
import networkx as nx
from networkx.algorithms import bipartite
from IPython.core.debugger import set_trace
#leemos todos los datos del fichero, recordar que el fichero actorMovies.csv fue modificado cambiando el nombre
#de The Rock y eliminando las comillas dobles
with open('datos' + param + '.csv') as csvfile:
    datos = list(csv.DictReader(csvfile, delimiter=';'))
    
with open('concap.csv') as csvCoordenadas:
    coordenadas = list(csv.DictReader(csvCoordenadas, delimiter=','))
```

```python
#Obtenemos la longitud y latitud del codigo de cada pais
myMap = {}
for c in coordenadas:
    if(c['CountryCode'] != "NULL"):
        myMap[c['CountryCode']] = (c['CapitalLongitude'], c['CapitalLatitude'])
```

```python
datos[0]
```

```python
node_list = set()
edge_list = set()
G = nx.DiGraph()
for elemento in datos:
    importer = elemento['Importer']
    exporter = elemento['Exporter']
    
    if((importer in myMap) and (exporter in myMap)):
    
        if importer not in node_list:
            node_list.add(importer)
            G.add_node(importer)
            G.nodes[importer]['longitud'] = myMap[importer][0]
            G.nodes[importer]['latitud'] = myMap[importer][1]
        if exporter not in node_list:
            node_list.add(exporter)
            G.add_node(exporter)
            G.nodes[exporter]['longitud'] = myMap[exporter][0]
            G.nodes[exporter]['latitud'] = myMap[exporter][1]
        if (importer, exporter) not in edge_list:
            edge_list.add((importer, exporter))
            G.add_edge(importer, exporter, weight = 1)
        else:
            G[importer][exporter]['weight'] += 1
            
print(node_list)
```

```python
#hacemos la lista de nodos para que gephi pueda leer el archivo correctamente
f=open("node_list_" + param + "_exportImport.csv","w", encoding = "utf-8")
f.write("ID\tLabel\tLongitud\tLatidud\n")
cont=1
mapa = {}
for n in G.nodes():
    f.write(str(cont) + "\t" + n + "\t"+ myMap[n][0] + "\t" + myMap[n][1] +"\n")
    mapa[n] = cont
    cont = cont + 1
f.close() 
```

```python
#hacemos la lista de aristas para que gephi pueda leer el archivo correctamente
f=open("edge_list_" + param + "_exportImport.csv","w", encoding = "utf-8")
f.write("Source\tTarget\tWeight\n")
for e in G.edges():
    f.write(str(mapa[e[0]]) + "\t" + str(mapa[e[1]]) + "\t" + str(G[str(e[0])][str(e[1])]['weight']) + "\n")
   # f.write(str(e[0]) + "\t" + str(e[1]) + "\n")
f.close()
```

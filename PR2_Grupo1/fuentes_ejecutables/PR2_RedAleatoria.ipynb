```python
import networkx as nx
import random
%matplotlib inline
```

```python
#generacion de la red aleatoria
def red_aleatoria(N, p):
    G = nx.Graph()
    G.add_nodes_from(range(N))
    for n1 in range(N) :
        for n2 in range(n1+1,N) :
            numAleat = random.random() #de 0 a 1 
            if (numAleat <= p) :
                G.add_edge(n1,n2)
                
    return G
```

```python
nodos = int(input("Introduce num nodos "))
probablidad = float(input("Introduce probabilidad "))
G = red_aleatoria(nodos, probablidad)
nx.draw(G)
```

```python
#hacemos la lista de nodos para que gephi pueda leer el archivo correctamente
f=open("node_list.csv","w", encoding = "utf-8")
f.write("ID\tLabel\n")
cont=0
mapa = {}
for n in G.nodes():
    f.write(str(cont) + "\t" + str(n) + "\n")
    mapa[n] = cont
    cont = cont + 1
f.close()  
```

```python
#hacemos la lista de aristas para que gephi pueda leer el archivo correctamente
f=open("edge_list.csv","w", encoding = "utf-8")
f.write("Source\tTarget\n")
for e in G.edges():
    f.write(str(mapa[e[0]]) + "\t" + str(mapa[e[1]]) + "\n")
f.close()
```
